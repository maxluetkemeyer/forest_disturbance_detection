```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv("../data/train.csv")
train_df.head()
```

```python
# group based on forest_id
grouped = train_df.sort_values("year").groupby("forest_id")
```

```python
# create sliding windows of size 3 years with shift 1 year
def create_sliding_windows(group, window_size=3):
    return group.rolling(window=window_size ).mean().shift(1)
# apply the function to each group
windows = grouped.apply(create_sliding_windows, window_size=3)
```
