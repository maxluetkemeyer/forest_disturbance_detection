# Example Notebook

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from numpy.typing import NDArray
```

```python
x: NDArray[np.float128] = np.array(object=[0.1, 0.25, 0.8], dtype=np.float128)
y: NDArray[np.float128] = np.array(object=[0.15, 0.4, 0.6], dtype=np.float128)
```

```python
# plot data with seaborn
df: pd.DataFrame = pd.DataFrame(data={
    'x': x,
    'y': y
})
sns.scatterplot(data=df, x='x', y='y')

# Set the range for x and y axes
plt.xlim(0, 1)
plt.ylim(0, 1)

# Show the plot
plt.show() # type: ignore
```

## Add a regression model to the notebook

```python
from sklearn.linear_model import LinearRegression

# Create a linear regression model
model: LinearRegression = LinearRegression()

# Fit the model using the data
model.fit(X=df[['x']], y=df['y']) # type: ignore

# Predict y values
df['y_pred'] = model.predict(X=df[['x']]) # type: ignore

# Plot the original data and the regression line
sns.scatterplot(data=df, x='x', y='y', label='Original Data')
sns.lineplot(data=df, x='x', y='y_pred', color='red', label='Regression Line')

# Set the range for x and y axes
plt.xlim(0, 1) # type: ignore
plt.ylim(0, 1) # type: ignore

# Show the plot
plt.show() # type: ignore
```
