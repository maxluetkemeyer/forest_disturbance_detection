```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt

train_plot = pd.read_csv("../../data/train.csv")
train_plot.head()

train_101 = train_plot[train_plot["forest_id"] == 101]
plt.figure(figsize=(10, 10))
plt.plot(train_101['year'], train_101['blue'])
plt.xlabel('Year')
plt.ylabel('Blue')
plt.title('Blue Values Over Years for Forest ID 101')
plt.show()
```
