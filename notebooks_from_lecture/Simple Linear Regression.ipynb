This notebook demonstrates how to fit a simple linear regression model for an artificial data set.

```python
# import the numpy package
import numpy as np
```

First, we generate an artificial data set with d=1 features and a target y. 

```python
# set "seed" such that the same "random" data set
# will be generated when executing this cell
np.random.seed(0)

# generate artificial dataset
N = 100
X = np.sort(1.5 * np.random.rand(N, 1), axis=0).reshape((N, 1))
y = np.sin(X) + 0.2*np.random.randn(N).reshape((N,1))
# y = 2 * X + 1 + 0.2*np.random.randn(N).reshape((N,1))
```

```python
import matplotlib.pyplot as plt

# plot the data
plt.scatter(X[:,0], y)
plt.xlabel('Feature (x)')
plt.ylabel('Target (y)')
```

Next, we define a function that, given some input data, computes the optimal coefficients for our linear regression model.

```python
def simple_linear_regression(X, y):
    """ 
    Computes an optimal solution for linear regression (1D case)
    
    Parameters
    ----------
    X : A numpy array containing the input variables
    y : A numpy array containing the output variables
    
    """
    
    # our method only works for d=1 features, i.e., the second
    # dimension of X must equal 1!
    assert X.shape[1] == 1
       
    N = len(X)
    
    # see Slide 31 of L1
    x_bar = np.mean(X)
    xy_bar = (1 / N) * (X*y).sum()    
    y_bar = np.mean(y)
    x_squared_bar = np.mean(X**2)
    
    w1 = (xy_bar - x_bar*y_bar)/(x_squared_bar - x_bar**2)
    w0 = (y_bar - w1*x_bar)
    
    return w0, w1
```

Let's apply the function defined above to get our coefficients ...

```python
w0, w1 = simple_linear_regression(X, y)
print("Optimal coefficients for simple linear regression model: w0={}, w1={}".format(w0, w1))
```

Finally, let us plot the data again along with our model (red line).

```python
plt.scatter(X, y)
xplot = np.linspace(0,1.5,100)
yplot = w0 + w1*xplot
plt.plot(xplot, yplot, 'r')
plt.xlabel('Feature (x)')
plt.ylabel('Target (y)')
```
